### plotting the graph

```python
# count plot for categorical cariables

categorical_df = df.select_dtypes(include=[object])
categorical_df.head()
```

```python
for i in categorical_df:
    plt.figure()  
    sns.countplot(x=i, data=categorical_df)
    plt.title(f'Count plot for {i}')
    plt.show()
```

```python
for i in categorical_df:
    for j in categorical_df:
        if i!=j:
            plt.figure()  
            sns.countplot(x=i, data=categorical_df,hue=j)
            plt.title(f'Count plot for {i}')
            plt.show()
```

```python
numeric_df.dtypes
```

```python
for i in numeric_df:
    plt.figure()
    sns.distplot(numeric_df[i],kde=True)
    plt.title(f'Count plot for {i}')
    plt.show()
```

```python
for i in numeric_df:
    for j in numeric_df:
        if i!=j and i!='Customer ID' and j!= "Customer ID":
            plt.figure()  
            sns.scatterplot(x=i,y=j, data=numeric_df,hue=categorical_df['Gender'])
            plt.title(f'Count plot for {i}')
            plt.show()
```

```python
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
